```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="darkgrid", color_codes=True)
```

```python
chm13_insert = pd.read_csv('/Users/ebiederstedt/Downloads/insertions_CHM13.no800.csv')
chm13_del = pd.read_csv('/Users/ebiederstedt/Downloads/deletions_CHM13.no800.csv')
korean_insert = pd.read_csv('/Users/ebiederstedt/Downloads/insertions_Korean.no800.csv')
korean_del = pd.read_csv('/Users/ebiederstedt/Downloads/deletions_Korean.no800.csv')
```

```python
sns.lmplot('NM', 'length', data=chm13_insert, fit_reg=False)
ax = plt.gca()
ax.set_title('CHM13 insertions: NM vs. length')
```

```python
print('chm13 insertions, ratio mean')
print(chm13_insert.ratio.mean())

print('chm13 deletions, ratio mean')
print(chm13_del.ratio.mean())

print('korean insertions, ratio mean')
print(korean_insert.ratio.mean())


print('korean deletions, ratio mean')
print(korean_del.ratio.mean())

```
