```r
library(ggplot2)
library(data.table)
```

```r
CHM1 = fread('/Users/ebiederstedt/GraphGenome_viz/data/NM_INS_DEL_CHM1.no800.csv')
CHM13 = fread('/Users/ebiederstedt/GraphGenome_viz/data/NM_INS_DEL_CHM13.no800.csv')
HG003 = fread('/Users/ebiederstedt/GraphGenome_viz/data/NM_INS_DEL_HG003.no800.csv')
HG004 = fread('/Users/ebiederstedt/GraphGenome_viz/data/NM_INS_DEL_HG004.no800.csv')
HX1 = fread('/Users/ebiederstedt/GraphGenome_viz/data/NM_INS_DEL_HX1.no800.csv')
Korean = fread('/Users/ebiederstedt/GraphGenome_viz/data/NM_INS_DEL_Korean.no800.csv')
NA19240 = fread('/Users/ebiederstedt/GraphGenome_viz/data/NM_INS_DEL_NA19240.no800.csv')
```

```r
CHM1[, assembly := 'CHM1']
CHM13[, assembly := 'CHM13']
HG003[, assembly := 'HG003']
HG004[, assembly := 'HG004']
HX1[, assembly := 'HX1']
Korean[, assembly := 'Korean']
NA19240[, assembly := 'NA19240']
```

```r
list_of_dts = list(CHM1, CHM13, HG003, HG004, HX1, Korean, NA19240)
```

```r
final = Reduce(function(...) merge(..., all = TRUE), list_of_dts)
```

```r
final = final[, c('INS_readLength', 'DEL_readLength', 'NM_INS_DEL', 'assembly')]
```

```r
final
```

```r
pp = ggplot(final, aes(assembly, INS_readLength, fill=assembly)) + geom_boxplot() + scale_y_continuous(limits=c(0,0.04)) +
     labs(title= "# of INS/read length", x="Assembly", y = "(# of INS)/(length)")  + theme(legend.position="none")


## + scale_fill_brewer(palette="Dark2") 

### primary only!!
print(pp)
```

```r
pp = ggplot(final, aes(assembly, DEL_readLength, fill=assembly)) + geom_boxplot() + scale_y_continuous(limits=c(0,0.04)) +
     labs(title= "# of DEL/read length", x="Assembly", y = "(# of DEL)/(length)")  + theme(legend.position="none")


## + scale_fill_brewer(palette="Dark2") 

### primary only!!
print(pp)
```

```r
pp = ggplot(final, aes(assembly, NM_INS_DEL, fill=assembly)) + geom_boxplot() + scale_y_continuous(limits=c(0,500)) +
     labs(title= "NM-INS-DEL", x="Assembly", y = "NM-INS-DEL")  + theme(legend.position="none")


## + scale_fill_brewer(palette="Dark2") 

### primary only!!
print(pp)
```
